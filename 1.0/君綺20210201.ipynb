```python
#爬取資料
def get_detail():
#  全域變數
    global brand
    global doctor_name
    global position
    global educationexperience
    global pic
#  (重複動作，取得醫生名﹑職位﹑教育經驗﹑照片)
    for doctor_names in driver.find_elements_by_xpath("//div[@class='doctor_name']//following-sibling::p[last()-1]"):
        doctor_name.append(doctor_names.text)    
    for positions in driver.find_elements_by_xpath("//div[@class='doctor']//following-sibling::ul[last()-1]"):
        position.append(positions.text)
    for educationexperiences in driver.find_elements_by_xpath("//div[@class='doctor']//following-sibling::ul[last()]"):
        educationexperience.append(educationexperiences.text)
    tempsrc=driver.find_elements_by_xpath("//li[contains(@name,'doctor')]//img")
    for ele in tempsrc:
        pics=ele.get_attribute("src")
        pic+=[pics]

#匯入套件
from selenium import webdriver
import time
from selenium.webdriver.common.action_chains import ActionChains
from bs4 import BeautifulSoup
import re
import pandas as pd
BRAND='君綺醫美'
#建立存取資料
brand=[]
doctor_name=[]
position=[]
educationexperience=[]
pic=[]
#開啟整形外科團隊
driver=webdriver.Chrome()
driver.get('https://www.jing-chi.com.tw')
driver.set_window_size(800,1000)
driver.find_element_by_xpath("//div[contains(@class,'nav_main_bar_logo')]//span[@class='navbar-toggler-icon']").click()
time.sleep(1)
driver.find_element_by_xpath("//*[@id='nav_about']").click()
driver.find_element_by_xpath("//*[@id='navbarNav']/ul/li[6]/ul/li[4]/span").click()
time.sleep(2)
driver.find_element_by_xpath("//*[@id='navbarNav']/ul/li[6]/ul/li[4]/ul/li[1]/a").click()
time.sleep(3)
#爬取整形外科團隊資料
get_detail()
#開啟微整美型團隊
driver.find_element_by_xpath("//div[contains(@class,'nav_main_bar_logo')]//span[@class='navbar-toggler-icon']").click()
time.sleep(1)
driver.find_element_by_xpath("//*[@id='nav_about']").click()
driver.find_element_by_xpath("//*[@id='navbarNav']/ul/li[6]/ul/li[4]/span").click()
time.sleep(2)
driver.find_element_by_xpath("//*[@id='navbarNav']/ul/li[6]/ul/li[4]/ul/li[2]/a").click()
time.sleep(3)
#爬取微整美型團隊資料
get_detail()
#計算資料，並新增相同比數的品牌名稱
for brands in range(len(doctor_name)):
    brand.append(BRAND) 
    
#整理爬取資料
def replacefunc(temp):
    temp = [w.replace('\n', '、') for w in temp]
replacefunc(doctor_name)
replacefunc(position)
replacefunc(educationexperience)
replacefunc(pic)

#執行程式
FILE_NAME = BRAND
today = time.strftime("%Y%m%d")
df=pd.DataFrame({'brand':brand,'doctor_name':doctor_name,'position':position,'education&experience':educationexperience,'pic':pic})
df.to_csv(f"{FILE_NAME}-{today}.csv",index=False)
```
